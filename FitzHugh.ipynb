```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = 0.2
b = 1
i = 0.7 # i = 0.7 or 0
e = 0.01
d = 0.5

x0 = [0.4, 0]
t = np.linspace(0, 350, 200)

def f(x, t, a, b, i, e, d):
    dx_dt=[-x[0]*(x[0] - a)*(x[0] - b) - x[1] + i, e * (x[0] - d*x[1])]
    return dx_dt

sol_x = odeint(f, x0, t, args=(a, b, i, e, d))

fig, axes = plt.subplots(1, 1, figsize=(12,4), dpi = 300)
axes.plot(t, sol_x[:, 0], label='v', alpha=0.5, marker = 'o', markersize=4)
axes.plot(t, sol_x[:, 1], label='w', alpha=0.5, marker ='s', markersize=4)
axes.legend(loc='best')
axes.grid(True, linestyle=(0, (1, 4)), linewidth=1)
plt.show()
```

```python
#intervall (for i = 0.7)
x_gebiet = np.arange(-0.5, 1.25,0.001)
y_gebiet = np.arange(0.55,0.9,0.001)
[X,Y]= np.meshgrid(x_gebiet,y_gebiet)

#system ('**' = '^')
xdot = -X*(X - a)*(X - b) - Y + i
ydot = e * (X - d*Y)

#colorizing
speed = np.sqrt(xdot**2+ydot**2)

#plot
fig, axes=plt.subplots(figsize=(15, 10), dpi = 300)
axes.streamplot(X,Y,xdot,ydot, color=speed, cmap='coolwarm', density=2)
axes.set_aspect('equal')
axes.set_xlabel('x')
axes.set_ylabel(r'$\partial_t x$')
axes.tick_params(left=True, bottom=True, top=True, right=True, direction='in')
axes.grid(True, linestyle=(0, (1, 4)), linewidth=1)
plt.show()
```
